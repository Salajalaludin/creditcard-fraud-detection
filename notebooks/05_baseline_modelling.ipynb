```python
# Menentukan root project dan menyiapkan library yang diperlukan.
from pathlib import Path
import subprocess, sys
import pandas as pd
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()

# Menjalankan pipeline training lengkap agar hasil notebook sama dengan CLI.
subprocess.run([sys.executable, str(ROOT / 'scripts' / 'train_baseline.py')], cwd=ROOT, check=True)
```

```python
# Membaca metrik yang sudah disimpan oleh training script.
metrics = pd.read_csv(ROOT / 'reports' / 'baseline_metrics.csv')

# Mengurutkan model berdasarkan split dan PR-AUC agar mudah dibandingkan.
metrics.sort_values(['split', 'pr_auc'], ascending=[True, False])
```
