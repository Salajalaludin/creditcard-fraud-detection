```python
# Menentukan root project dan membaca artefak tanpa melatih ulang model.
from pathlib import Path
import pandas as pd
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
recommendations = pd.read_csv(ROOT / 'reports' / 'threshold_recommendations.csv')
recommendations.loc[recommendations['strategy'].str.startswith('business_'), ['strategy', 'threshold', 'precision', 'recall', 'alerts', 'total_cost']]
```

```python
# Menampilkan transaksi dengan prioritas investigasi tertinggi.
queue = pd.read_csv(ROOT / 'data' / 'predictions' / 'investigation_queue.csv')
queue.head(20)
```

```python
# Meringkas jumlah alert dan nominal berdasarkan risk level.
queue.groupby('risk_level', observed=False).agg(alerts=('transaction_id', 'size'), amount_total=('Amount', 'sum'), average_score=('risk_score', 'mean'))
```
